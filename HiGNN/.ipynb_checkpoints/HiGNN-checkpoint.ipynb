# HiGNN: hierarchical molecular property prediction
## A GPU-oriented reproduction of Zhu *et al.*

This notebook explains and reproduces the central ideas of **HiGNN: A Hierarchical Informative Graph Neural Network for Molecular Property Prediction Equipped with Feature-Wise Attention** by Weimin Zhu *et al.*

- [Authors' HiGNN repository](https://github.com/idrugLab/HiGNN)
- [Paper title search on PubMed](https://pubmed.ncbi.nlm.nih.gov/?term=HiGNN%3A+A+Hierarchical+Informative+Graph+Neural+Network)

The reproduction builds an atom graph and a BRICS-derived fragment graph, learns both resolutions, and combines them using feature-wise attention. MoleculeNet inputs, fragment metadata, scaffold splits, and predictions persist under `data/HiGNN/`; the best reusable checkpoint is written to `models/HiGNN/`.

## 1. The hierarchy

A conventional molecular GNN updates atom states along covalent bonds. HiGNN adds a second, chemically coarser view. BRICS rules cut retrosynthetically meaningful bonds and partition the molecule into fragments. Atom states are pooled into fragment states, which then communicate across the cut bonds.

For atom $v$ and fragment $k$, the schematic updates are

$$h_v^{(l+1)}=\operatorname{GNN}_{atom}(h_v^{(l)},\{h_u,e_{uv}\}),$$
$$s_k^{(l+1)}=\operatorname{GNN}_{frag}(s_k^{(l)},\{s_j,c_{kj}\}).$$

This makes local functional-group information explicit without discarding the original atom graph.

## 2. Feature-wise attention

After pooling atom and fragment representations into $g_a$ and $g_f$, a vector gate chooses the useful resolution separately for every latent feature:

$$\alpha=\sigma(\mathrm{MLP}([g_a\Vert g_f])),\qquad g=\alpha\odot g_a+(1-\alpha)\odot g_f.$$

Unlike a single scalar attention weight, this can prefer atom-scale evidence for some channels and fragment-scale evidence for others. The gate is exposed later for interpretation.

```python
import json, math, os, random, time
from pathlib import Path
import numpy as np
import pandas as pd
import torch
from torch import nn
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.datasets import MoleculeNet
from torch_geometric.loader import DataLoader
from torch_geometric.nn import GINEConv, global_mean_pool
from rdkit import Chem
from rdkit.Chem import BRICS
from rdkit.Chem.Scaffolds import MurckoScaffold
import matplotlib.pyplot as plt

SEED=42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
USE_CUDA=torch.cuda.is_available(); device=torch.device('cuda' if USE_CUDA else 'cpu')
if USE_CUDA:
    torch.cuda.manual_seed_all(SEED)
    torch.set_float32_matmul_precision('high')
    torch.backends.cudnn.benchmark=True
PROJECT_ROOT=Path.cwd().resolve()
if PROJECT_ROOT.name=='HiGNN': PROJECT_ROOT=PROJECT_ROOT.parent
DATA_DIR=PROJECT_ROOT/'data'/'HiGNN'; MODEL_DIR=PROJECT_ROOT/'models'/'HiGNN'
DATA_DIR.mkdir(parents=True,exist_ok=True); MODEL_DIR.mkdir(parents=True,exist_ok=True)
print(f'PyTorch={torch.__version__}; device={device}; data={DATA_DIR}')
```

## 3. MoleculeNet task and scaffold split

ESOL is the default compact regression benchmark. `FreeSolv` and `Lipo` are also regression tasks; `BBBP`, `BACE`, and `HIV` are binary classification tasks. A Bemis–Murcko scaffold split is stricter than a random split because related molecular cores remain together. The split is frozen on disk for exact reruns.

```python
DATASET_NAME='ESOL'  # ESOL, FreeSolv, Lipo, BBBP, BACE, or HIV
TASKS={'ESOL':'regression','FreeSolv':'regression','Lipo':'regression',
       'BBBP':'classification','BACE':'classification','HIV':'classification'}
if DATASET_NAME not in TASKS: raise ValueError(f'Unsupported task: {DATASET_NAME}')
TASK_TYPE=TASKS[DATASET_NAME]
dataset=MoleculeNet(root=DATA_DIR/'MoleculeNet',name=DATASET_NAME)
smiles=[str(dataset[i].smiles) for i in range(len(dataset))]

def scaffold_of(s):
    return MurckoScaffold.MurckoScaffoldSmiles(smiles=s,includeChirality=True)
def make_scaffold_split(strings):
    groups={}
    for i,s in enumerate(strings): groups.setdefault(scaffold_of(s),[]).append(i)
    groups=sorted(groups.values(),key=lambda x:(-len(x),x[0]))
    limits=(int(.8*len(strings)),int(.9*len(strings))); parts=[[],[],[]]
    for group in groups:
        destination=0 if len(parts[0])+len(group)<=limits[0] else (1 if len(parts[1])+len(group)<=limits[1]-limits[0] else 2)
        parts[destination].extend(group)
    return {k:torch.tensor(sorted(v),dtype=torch.long) for k,v in zip(('train','valid','test'),parts)}
SPLIT_FILE=DATA_DIR/f'{DATASET_NAME}_scaffold_seed{SEED}.pt'
splits=torch.load(SPLIT_FILE,weights_only=True) if SPLIT_FILE.exists() else make_scaffold_split(smiles)
if not SPLIT_FILE.exists(): torch.save(splits,SPLIT_FILE)
print(DATASET_NAME,TASK_TYPE,{k:len(v) for k,v in splits.items()})
```

## 4. Persistent BRICS fragment graphs

RDKit finds BRICS bonds. Removing those bonds gives connected atom components; every component is a fragment node, and every cut becomes an undirected fragment edge. This preprocessing is deterministic and cached so training epochs perform only tensor work.

```python
HIERARCHY_FILE=DATA_DIR/f'{DATASET_NAME}_brics_hierarchy.pt'
def brics_hierarchy(s):
    mol=Chem.MolFromSmiles(s)
    if mol is None: raise ValueError(f'Invalid SMILES: {s}')
    cuts={tuple(sorted(pair)) for pair,_ in BRICS.FindBRICSBonds(mol)}
    adjacency=[[] for _ in range(mol.GetNumAtoms())]
    for bond in mol.GetBonds():
        a,b=bond.GetBeginAtomIdx(),bond.GetEndAtomIdx()
        if tuple(sorted((a,b))) not in cuts: adjacency[a].append(b); adjacency[b].append(a)
    atom2frag=torch.full((mol.GetNumAtoms(),),-1,dtype=torch.long); n_frag=0
    for start in range(mol.GetNumAtoms()):
        if atom2frag[start]>=0: continue
        stack=[start]; atom2frag[start]=n_frag
        while stack:
            u=stack.pop()
            for v in adjacency[u]:
                if atom2frag[v]<0: atom2frag[v]=n_frag; stack.append(v)
        n_frag+=1
    frag_edges=[]
    for a,b in cuts:
        fa,fb=int(atom2frag[a]),int(atom2frag[b]); frag_edges.extend([(fa,fb),(fb,fa)])
    edge_index=torch.tensor(frag_edges,dtype=torch.long).t().contiguous() if frag_edges else torch.empty((2,0),dtype=torch.long)
    return {'atom2frag':atom2frag,'frag_edge_index':edge_index,'num_frags':n_frag}
if HIERARCHY_FILE.exists():
    hierarchy=torch.load(HIERARCHY_FILE,weights_only=True)
else:
    hierarchy=[]
    for i,s in enumerate(smiles):
        item=brics_hierarchy(s)
        if len(item['atom2frag'])!=dataset[i].num_nodes: raise RuntimeError(f'Atom-order mismatch at {i}')
        hierarchy.append(item)
    torch.save(hierarchy,HIERARCHY_FILE)
if len(hierarchy)!=len(dataset): raise ValueError('Stale hierarchy cache; delete it and rerun.')
print(f'Cached {len(hierarchy):,} hierarchies; mean fragments={np.mean([x["num_frags"] for x in hierarchy]):.2f}')
```

```python
class HierData(Data):
    def __inc__(self,key,value,*args,**kwargs):
        if key in ('atom2frag','frag_edge_index'): return self.num_frags
        return super().__inc__(key,value,*args,**kwargs)
class HierDataset(torch.utils.data.Dataset):
    def __init__(self,base,meta,indices): self.base,self.meta,self.indices=base,meta,indices.tolist()
    def __len__(self): return len(self.indices)
    def __getitem__(self,j):
        i=self.indices[j]; data=self.base[i]
        return HierData(**data.to_dict(),**self.meta[i])
BATCH_SIZE=256 if USE_CUDA else 64
NUM_WORKERS=min(8,os.cpu_count() or 1) if USE_CUDA else 0
loader_args=dict(batch_size=BATCH_SIZE,num_workers=NUM_WORKERS,pin_memory=USE_CUDA,persistent_workers=NUM_WORKERS>0)
loaders={k:DataLoader(HierDataset(dataset,hierarchy,v),shuffle=(k=='train'),**loader_args) for k,v in splits.items()}
train_y=torch.cat([dataset[int(i)].y.view(-1) for i in splits['train']]).float()
Y_MEAN=train_y[torch.isfinite(train_y)].mean() if TASK_TYPE=='regression' else torch.tensor(0.)
Y_STD=train_y[torch.isfinite(train_y)].std().clamp_min(1e-8) if TASK_TYPE=='regression' else torch.tensor(1.)
atom_sizes=[max(int(dataset[i].x[:,j].max()) for i in range(len(dataset)))+1 for j in range(dataset.num_node_features)]
edge_sizes=[max(int(dataset[i].edge_attr[:,j].max()) if dataset[i].edge_attr.numel() else 0 for i in range(len(dataset)))+1 for j in range(dataset.num_edge_features)]
print('atom vocab',atom_sizes,'bond vocab',edge_sizes,'target mean/std',float(Y_MEAN),float(Y_STD))
```

## 5. Hierarchical informative GNN

Categorical atom and bond fields receive separate embeddings. Edge-aware GINE layers operate on atoms. Atom states are pooled into BRICS fragments, followed by fragment-level message passing. The feature-wise gate fuses the two graph summaries before prediction. Residual connections, normalization, and dropout make the deeper model practical on a GPU cluster.

```python
HIDDEN=256; ATOM_LAYERS=4; FRAG_LAYERS=2; DROPOUT=.15
class FieldEncoder(nn.Module):
    def __init__(self,sizes,hidden):
        super().__init__(); self.embeddings=nn.ModuleList([nn.Embedding(size,hidden) for size in sizes])
        for emb in self.embeddings: nn.init.xavier_uniform_(emb.weight)
    def forward(self,x): return torch.stack([emb(x[:,i]) for i,emb in enumerate(self.embeddings)]).sum(0)
def gine(hidden):
    return GINEConv(nn.Sequential(nn.Linear(hidden,2*hidden),nn.SiLU(),nn.Linear(2*hidden,hidden)),train_eps=True)
class HiGNN(nn.Module):
    def __init__(self):
        super().__init__(); self.atom_encoder=FieldEncoder(atom_sizes,HIDDEN); self.bond_encoder=FieldEncoder(edge_sizes,HIDDEN)
        self.atom_convs=nn.ModuleList([gine(HIDDEN) for _ in range(ATOM_LAYERS)])
        self.atom_norms=nn.ModuleList([nn.LayerNorm(HIDDEN) for _ in range(ATOM_LAYERS)])
        self.cut_edge=nn.Parameter(torch.zeros(HIDDEN)); self.frag_convs=nn.ModuleList([gine(HIDDEN) for _ in range(FRAG_LAYERS)])
        self.frag_norms=nn.ModuleList([nn.LayerNorm(HIDDEN) for _ in range(FRAG_LAYERS)])
        self.attention=nn.Sequential(nn.Linear(2*HIDDEN,HIDDEN),nn.SiLU(),nn.Linear(HIDDEN,HIDDEN))
        self.predictor=nn.Sequential(nn.Linear(HIDDEN,HIDDEN),nn.SiLU(),nn.Dropout(DROPOUT),nn.Linear(HIDDEN,1))
    def forward(self,batch,return_attention=False):
        h=self.atom_encoder(batch.x); edge_attr=self.bond_encoder(batch.edge_attr)
        for conv,norm in zip(self.atom_convs,self.atom_norms):
            h=norm(h+F.dropout(conv(h,batch.edge_index,edge_attr),DROPOUT,self.training))
        atom_graph=global_mean_pool(h,batch.batch)
        frag=global_mean_pool(h,batch.atom2frag)
        cut_attr=self.cut_edge.expand(batch.frag_edge_index.shape[1],-1)
        for conv,norm in zip(self.frag_convs,self.frag_norms):
            frag=norm(frag+F.dropout(conv(frag,batch.frag_edge_index,cut_attr),DROPOUT,self.training))
        frag_batch=torch.repeat_interleave(torch.arange(batch.num_graphs,device=h.device),batch.num_frags)
        frag_graph=global_mean_pool(frag,frag_batch)
        alpha=torch.sigmoid(self.attention(torch.cat([atom_graph,frag_graph],dim=-1)))
        prediction=self.predictor(alpha*atom_graph+(1-alpha)*frag_graph).squeeze(-1)
        return (prediction,alpha) if return_attention else prediction
model=HiGNN().to(device)
print(f'parameters={sum(p.numel() for p in model.parameters()):,}')
```

## 6. GPU training and reusable checkpoint

Regression targets are standardized using only the training split. Classification uses logits and ROC-AUC; regression reports RMSE and MAE in the original units. Automatic mixed precision, pinned-memory loading, fused AdamW on CUDA, early stopping, and a learning-rate scheduler are enabled. A compatible checkpoint is loaded instead of retraining.

```python
EPOCHS=150; PATIENCE=25; FORCE_RETRAIN=False
CHECKPOINT=MODEL_DIR/f'hignn_{DATASET_NAME.lower()}.pt'
CONFIG={'dataset':DATASET_NAME,'task':TASK_TYPE,'hidden':HIDDEN,'atom_layers':ATOM_LAYERS,'frag_layers':FRAG_LAYERS,'dropout':DROPOUT,'seed':SEED,'atom_sizes':atom_sizes,'edge_sizes':edge_sizes}
def binary_auc(y,score):
    pos=score[y==1]; neg=score[y==0]
    if len(pos)==0 or len(neg)==0: return float('nan')
    return float(((pos[:,None]>neg[None,:]).float()+.5*(pos[:,None]==neg[None,:]).float()).mean())
@torch.no_grad()
def evaluate(loader,with_rows=False):
    model.eval(); ys=[]; preds=[]; row_ids=[]
    for batch in loader:
        batch=batch.to(device,non_blocking=True); pred=model(batch); y=batch.y.view(-1).float(); mask=torch.isfinite(y)
        ys.append(y[mask].cpu()); preds.append(pred[mask].cpu())
    y=torch.cat(ys); pred=torch.cat(preds)
    if TASK_TYPE=='regression':
        pred=pred*Y_STD+Y_MEAN; metrics={'rmse':float(torch.sqrt(F.mse_loss(pred,y))),'mae':float(F.l1_loss(pred,y))}
    else:
        pred=pred.sigmoid(); metrics={'roc_auc':binary_auc(y,pred),'accuracy':float(((pred>=.5)==y.bool()).float().mean())}
    return metrics,y,pred
history={'train':[],'valid':[]}; best=float('inf') if TASK_TYPE=='regression' else -float('inf'); stale=0
if CHECKPOINT.exists() and not FORCE_RETRAIN:
    ckpt=torch.load(CHECKPOINT,map_location=device,weights_only=True)
    if ckpt['config']!=CONFIG: raise ValueError('Checkpoint configuration differs; set FORCE_RETRAIN=True.')
    model.load_state_dict(ckpt['model']); history=ckpt.get('history',history); print('Loaded',CHECKPOINT)
else:
    optimizer=torch.optim.AdamW(model.parameters(),lr=2e-4,weight_decay=1e-5,fused=USE_CUDA)
    scheduler=torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer,mode='min' if TASK_TYPE=='regression' else 'max',factor=.5,patience=8)
    scaler=torch.amp.GradScaler('cuda',enabled=USE_CUDA); started=time.time()
    for epoch in range(1,EPOCHS+1):
        model.train(); total=count=0
        for batch in loaders['train']:
            batch=batch.to(device,non_blocking=True); y=batch.y.view(-1).float(); mask=torch.isfinite(y); optimizer.zero_grad(set_to_none=True)
            with torch.amp.autocast('cuda',enabled=USE_CUDA):
                pred=model(batch); target=(y-Y_MEAN.to(device))/Y_STD.to(device) if TASK_TYPE=='regression' else y
                loss=F.mse_loss(pred[mask],target[mask]) if TASK_TYPE=='regression' else F.binary_cross_entropy_with_logits(pred[mask],target[mask])
            scaler.scale(loss).backward(); scaler.unscale_(optimizer); nn.utils.clip_grad_norm_(model.parameters(),5.)
            scaler.step(optimizer); scaler.update(); total+=loss.item()*int(mask.sum()); count+=int(mask.sum())
        valid,_,_=evaluate(loaders['valid']); key=valid['rmse'] if TASK_TYPE=='regression' else valid['roc_auc']
        history['train'].append(total/max(1,count)); history['valid'].append(key); scheduler.step(key)
        improved=key<best if TASK_TYPE=='regression' else key>best
        if improved:
            best=key; stale=0; torch.save({'model':model.state_dict(),'config':CONFIG,'history':history,'y_mean':Y_MEAN,'y_std':Y_STD,'best_valid':best},CHECKPOINT)
        else: stale+=1
        if epoch==1 or epoch%10==0: print(f'{epoch:03d} loss={total/max(1,count):.4f} valid={valid} lr={optimizer.param_groups[0]["lr"]:.2g}')
        if stale>=PATIENCE: print('Early stopping at',epoch); break
    print(f'Training time={(time.time()-started)/60:.1f} min')
    ckpt=torch.load(CHECKPOINT,map_location=device,weights_only=True); model.load_state_dict(ckpt['model'])
test_metrics,test_y,test_pred=evaluate(loaders['test'])
print('Test',json.dumps(test_metrics,indent=2))
```

```python
prediction_file=DATA_DIR/f'{DATASET_NAME}_test_predictions.csv'
pd.DataFrame({'dataset_index':splits['test'].numpy(),'smiles':[smiles[int(i)] for i in splits['test']],
              'target':test_y.numpy(),'prediction':test_pred.numpy()}).to_csv(prediction_file,index=False)
batch=next(iter(loaders['test'])).to(device); model.eval()
with torch.no_grad(): _,alpha=model(batch,return_attention=True)
fig,axes=plt.subplots(1,2,figsize=(11,3.5))
axes[0].plot(history['train'],label='train objective'); axes[0].set(xlabel='epoch',title='Optimization'); axes[0].legend()
axes[1].hist(alpha.cpu().flatten().numpy(),bins=30); axes[1].axvline(.5,color='black',ls='--'); axes[1].set(xlabel='atom-resolution gate',title='Feature-wise attention')
plt.tight_layout(); print('Predictions saved to',prediction_file)
```

## 7. Reading the reproduction responsibly

### What is reproduced

- a fine atom graph and chemically motivated coarse fragment graph;
- information propagation at both resolutions;
- feature-wise attention between molecular representations;
- scaffold-split property prediction and attention inspection.

### Deliberate implementation choices

This is a clean PyTorch Geometric reproduction of the paper's central mechanism, not a byte-for-byte port of its original environment. GINE layers, residual normalization, BRICS cuts, and the optimizer schedule are explicit modern choices. Exact paper tables require the paper's dataset versions, split protocol, hyperparameter search, and repeated seeds.

Attention is a useful diagnostic but is not automatically a causal explanation. Scaffold-split performance can also vary substantially with a single split. For a publishable comparison, run multiple seeds, preserve every split, report confidence intervals, compare atom-only and fragment-only ablations, and check whether attention is stable across retraining.
